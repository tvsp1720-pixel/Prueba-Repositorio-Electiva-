```python
# Grafica Proyecto
import matplotlib.pyplot as plt

# Datos
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Crear la gráfica
plt.plot(x, y, color='blue', marker='o', linestyle='--')

# Etiquetas
plt.title("Mi primera gráfica")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")

# Mostrar
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. Datos
lenguajes = ['Python', 'JavaScript', 'C++', 'Java']
popularidad = [85, 70, 45, 60]
colores = ['skyblue', 'gold', 'lightcoral', 'lightgreen']

# 2. Crear la gráfica
plt.bar(lenguajes, popularidad, color=colores, edgecolor='black')

# 3. Personalización (el "maquillaje")
plt.title('Popularidad de Lenguajes', fontsize=14)
plt.xlabel('Lenguajes de Programación')
plt.ylabel('Puntaje de Uso')
plt.grid(axis='y', linestyle='--', alpha=0.7) # Rejilla solo horizontal

# 4. Mostrar el resultado
plt.show()
```
