```python
# Grafica Proyecto
import matplotlib.pyplot as plt

# Datos
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Crear la gráfica
plt.plot(x, y, color='blue', marker='o', linestyle='--')

# Etiquetas
plt.title("Mi primera gráfica")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")

# Mostrar
plt.show()
```
